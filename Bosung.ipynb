```python
n=list(map(int,input("Nhap 10 so nguyen:", ).split() ))
soduong=0
soam=0
sobangkhong=0
for i in n:
  if i>0:
    soduong+=1
  elif i<0:
    soam+=1
  else :
    sobangkhong+=1
print("Tong so nguyen duong ",soduong)
print("Tong so nguyen am", soam)
print("Tong so bang khong", sobangkhong)


```

```python
n = int(input("Nhap so de kiem tra: "))

if n < 2:
    print(n, "khong phai la so nguyen to")
else:
    so = True
    for i in range(2, n):
        if n % i == 0:
            so = False
            break

    if so:
        print(n, "la so nguyen to")
    else:
        print(n, "khong la so nguyen to")
```
