```python
n=float(input())
if n>=8.5:
  print("Xuat sac")
elif 7.0<=n<=8.5:
  print("Gioi")
elif 5.5<=n<=7.0:
  print("Kha")
elif 4.0<=n<=5.5:
  print("Trung Binh")
else:
  print("Yeu")
```
